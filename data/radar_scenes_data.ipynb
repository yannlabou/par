## Extraction of data from the Radar Scenes dataset

The scene from the dataset extracted using https://pypi.org/project/radar-scenes/



### General idea
Extract x and y measurements of one agent

````python
import h5py
import numpy as np
import pandas as pd
f = h5py.File('RadarScenes/data/sequence_1/radar_data.h5', 'r')
f.keys()
````

````python
radar_data_raw = f.get('radar_data')
radar_data_array = np.array(radar_data_raw)
radar_data_time_series = pd.DataFrame(radar_data_array)
print(radar_data_time_series)
````

==> Now we have measurements from different sensors of the same agent!!
