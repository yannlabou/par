# Bayesian Filter

```python
import pandas as pd

sensor_data = pd.read_csv('sensor_data.csv', delimiter=',')
bayesian_beliefs = pd.read_csv('bayesian_beliefs.csv', delimiter=';')
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.plot(sensor_data["# ground_truth_x"], sensor_data[" ground_truth_y"], 'or')

#plt.plot(x, y, 'red')
bayesian_beliefs['diffs_y-'] = sensor_data[" ground_truth_y"] - (sensor_data[" ground_truth_y"] - sensor_data[" ground_truth_y"]*bayesian_beliefs['y_belief'])
bayesian_beliefs['diffs_y+'] = sensor_data[" ground_truth_y"] + (sensor_data[" ground_truth_y"] - sensor_data[" ground_truth_y"]*bayesian_beliefs['y_belief'])

plt.fill_between(sensor_data["# ground_truth_x"], bayesian_beliefs['diffs_y-'][0:100], bayesian_beliefs['diffs_y+'][0:100],
                 color='gray', alpha=0.2)

plt.show()
```
